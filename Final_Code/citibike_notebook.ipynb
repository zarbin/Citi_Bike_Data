```python
# Import Libraries and Dependencies
import pandas as pd
import os
import glob
```

```python
# use glob to get all the csv files 
# in the folder
path = os.getcwd()
csv_files = glob.glob(os.path.join(path, "../Citibike_Tripdata/*.csv"))
```

```python
data = pd.DataFrame()

for f in csv_files:
    csv = pd.read_csv(f)
    data = pd.concat([data, csv])
    print('Loaded file:', f)
```
